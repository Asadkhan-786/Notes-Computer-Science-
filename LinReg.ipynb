```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

```

```python
# get the data
advertising = pd.read_csv('/content/Advertising.csv', index_col=0)
```

```python
advertising.head()
```

```python
X = advertising['TV']
Y = advertising['Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
```

```python
pd.DataFrame(X_train)
```

```python
##  Linear Regression Fitting example
model = LinearRegression()
model.fit(pd.DataFrame(X_train), pd.DataFrame(Y_train))
#
## prediction example
y_test_predict = model.predict(pd.DataFrame(X_test))
#
## model evaluation
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)
print(rmse,r2)
```

```python
advertising.describe()
```

```python
plt.scatter(X_test.values, Y_test.values,color='b')
plt.scatter(X_test.values, y_test_predict,color='r')
plt.show()
```
